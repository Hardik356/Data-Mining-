**Pass an argument**

Passing an argument refers to providing a value or expression to a function or method when it is called. Functions in Python can be defined with parameters, which act as placeholders for the values that will be supplied when the function is invoked.

````python
def myFunction(Name):
  print("Hi, "+ Name +" I am inside the function")

myFunction("Natash")
````

**Pass two arguments**

In Python, you can define a function with multiple parameters to pass two or more values.

````python
def myFunction(Name, Age):
  print("Hi, "+Name +" I am inside the function")
  print("I know your age is", Age, ", "+ Name)

myFunction("Natash",32)
````

**Default parameter value(s)**

Default parameter values in Python allow you to assign default values to function parameters. When a function is called, if an argument is not provided for a parameter with a default value, the default value is used. This provides flexibility and makes it possible to call a function with fewer arguments, relying on the default values for the rest.

````python
def myFunction(Name="noName", Age=99999):
  print("Hi, "+Name +" I am inside the function")
  print("I know your age is", Age, ", "+ Name)

myFunction()

myFunction("Natash",32)
````

**Factorial**

The factorial of a non-negative integer is the product of all positive integers less than or equal to that integer. It is denoted by the symbol $!$.

For example, the factorial of 5 (written as $5!$) is calculated as follows:

$5!=5×4×3×2×1=120$

In general, the factorial of a non-negative integer $n$ is expressed as:

$n!=n×(n−1)×(n−2)×…×2×1$

````python
n=int(input("Enter an integer: "))
n1=n
if n == 0 or n == 1:
  factorial = 1
else:
  factorial = 1
  while(n > 1):
    factorial *= n #factorial=factorial*n
    n -= 1

print ("Factorial of ", n1, "is", factorial)
````

**Factorial using a function**

````python
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        factorial = 1
        while(n > 1):
            factorial *= n
            n -= 1
        return factorial

a=int(input("Enter an integer: "))
print ("Factorial of ", a, "is", fact(a))
````

**Recursion**

A recursive function is a function that calls itself during its execution. In other words, the function solves a smaller instance of the same problem as part of its computation. Recursion is a powerful programming technique commonly used to break down complex problems into simpler, more manageable subproblems.

````python
def fact(x):
  if x == 0 or x == 1:
    return 1
  else:
    return x*fact(x-1)

a=int(input("Enter an integer: "))
print ("Factorial of ", a, "is", fact(a))
````

**Advanced approach**

````python
def fact(x):
  return 1 if (x == 0 or x == 1) else x*fact(x-1)

a=int(input("Enter an integer: "))
print ("Factorial of ", a, "is", fact(a))
````

**Arbitrary Arguments** use * before the argument: *args  ---- It means that the number of arguments is unknown

In Python, you can use ***args** in the parameter list of a function definition to allow it to accept an arbitrary number of positional arguments. The term "args" is just a convention; you can use any name you prefer, but ***args** is widely used and recognised. For example:

```
def print_arguments(*args):
    for arg in args:
        print(arg)

print_arguments(1, 'hi', [3, 4], 'there')

```



````python
def myList(*Name):
  for x in Name:
    print(x)

myList("Adam", "Mike", "Paul","Eric")
print("\n")
myList("Adam", "Mike", "Paul")
````

**Single Asterisk (*):**

Used in function definition to collect all positional arguments into a tuple.
Used in function call to unpack the elements of an iterable (like a list or tuple) as positional arguments.

````python
def example_function(*args):
    print(args)

example_function(1, 2, 3)
````

**Double Asterisk** (\**):

Used in function definition to collect all keyword arguments into a dictionary.
Used in function call to unpack the key-value pairs of a dictionary as keyword arguments.

````python
def example_function(**kwargs):
    print(kwargs)

example_function(a=1, b=2, c=3)
````

**Both Single  and Double Asterisks**  

````python
def example_function(*args, **kwargs):
    print("Positional Arguments (args):", args)
    print("Keyword Arguments (kwargs):", kwargs)

# Example Call
example_function(1, 2, 3, a=4, b=5)
````

**Python function: Lambda**

In Python, a lambda function is a concise way to create anonymous functions, also known as lambda expressions. Lambda functions are defined using the lambda keyword, followed by a list of parameters, a colon, and an expression. The syntax is as follows:

```
lambda parameters: expression

```



**Note: lambda any_number_of_arguments: only_one_expression**

````python
# Area of the Circle
r=float(input("Enter the radius:"))
Area = lambda r : 3.14*r*r
print("The area of the circle is: ",Area(r))
````

**Factorial using the lambda function**

````python
fact = lambda x: 1 if (x == 0 or x == 1) else x*fact(x-1)

n=int(input("Enter an integer:"))

print ("Factorial of ", n, "is", fact(n))
````

**Defining a Closure Function**

A closure is a function object that has access to variables in its lexical scope, even when the function is called outside that scope. Closures allow functions to retain information about the environment in which they were created. This behaviour is possible because functions in Python are first-class citizens, meaning they can be passed around and returned as values.

````python
def outerFunction(msg):
    def innerFunction():
        print(msg)
    return innerFunction

call=outerFunction("Hi from the inner function!")
call()
````

````python
def outer_function(x):
    # Inner function (closure)
    def inner_function(y):
        return x + y
    return inner_function

# Create closures
closure1 = outer_function(10)
closure2 = outer_function(20)

print(closure1)
print(closure2)

# Call closures
result1 = closure1(5)  # Output: 15 (10 + 5)
result2 = closure2(5)  # Output: 25 (20 + 5)

print(result1)
print(result2)
````
